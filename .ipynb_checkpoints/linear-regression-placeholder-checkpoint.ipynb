```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
"""
   직접 훈련 데이터를 미리 작성하지 않고, 실행 도중에 넘겨주기 위해 텐서플로우의 placeholder를 사용한다.
   2번째 인자로 shape=[None]을 던져주는데, 그 이유는 어떤 개수의 모양으로 넘겨줘도 상관 없게 만들기 위함이다.
"""
X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])

# make W and b for H(x) = Wx + b
W = tf.Variable(tf.random.normal([1]), name='weight')
b = tf.Variable(tf.random.normal([1]), name='bias')
```

```python
# Our hypothesis XW + b
hypothesis = X*W + b
```

```python
# const/loss function => 차이를 평균 내준다.
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
"""
    cost를 최소화 한다!
    텐서플로우가 알아서 최소 cost를 가지는 W와 b를 조정한다.
    이 부분은 텐서플로우 라이브러리가 알아서 해주는 magic이라 생각하자.
"""
optimizer = tf.compat.v1.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

```python
# make session
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run( [cost, W, b, train], feed_dict={X: [1, 2, 3, 4, 5], Y: [2.1, 3.1, 4.1, 5.1, 6.1]} )
    
    if step%20 == 0:
        print(step, cost_val, W_val, b_val)
```

```python
"""
    직관적으로 W=1, b=1.1쯤 나올 것으로 예상된다.
    이렇게 훈련을 거친 hypothesis를 테스트 해본다.
"""
print(sess.run(hypothesis, feed_dict={X: [5]}))
print(sess.run(hypothesis, feed_dict={X: [2.5]}))
print(sess.run(hypothesis, feed_dict={X: [1.5, 3.5]}))
```
